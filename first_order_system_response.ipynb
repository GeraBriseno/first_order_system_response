```python
import math
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = [i*.1 for i in range(1,31)]   #time
cT = 1                            #time constant
```

Linear First Order Systems  https://www.tutorialspoint.com/control_systems/control_systems_response_first_order.htm

```python
y = []
y2 = []
y3 = []
y4 = []
for i in t:
    
    y.append(1-math.exp(-(i/cT)))                                 #step
    y2.append(i-cT+(cT*(math.exp(-(i/cT)))))                      #ramp
    y3.append((1/cT)*(math.exp(-i/cT)))                           #impulse
    y4.append((i**2/2)-(cT*i)+(cT**2)-((cT**2)*math.exp(-i/cT)))  #parabola
```

```python
plt.plot(t,y,label="step")
plt.plot(t,y2,label="ramp")
plt.plot(t,y3,label="impulse")
plt.plot(t,y4,label="parabola")
plt.legend()
```
